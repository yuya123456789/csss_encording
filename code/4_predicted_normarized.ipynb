```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import os

# =========================================================
# 一括処理したいモデル名
# =========================================================
model_list = ["w2v", "llama", "llama3", "gpt2", "gpt_oss", "gte"]
story_train="bronx"
story_test="piemanpni"
# =========================================================
# ベースの入出力パス
# =========================================================
base_dir = "../predicted_all_csv"

# =========================================================
# 1 モデルごとにループ処理
# =========================================================
for model_name in model_list:

    input_csv  = f"{base_dir}/{model_name}_{story_train}_to_{story_test}_predicted.csv"
    output_csv = f"{base_dir}/{model_name}_{story_train}_to_{story_test}_predicted_scaled.csv"

    if not os.path.exists(input_csv):
        print(f"⚠ 入力CSVが見つかりません: {input_csv}")
        continue

    print(f"--- {model_name}: スケーリング開始 ---")

    # -------------------------------------------------------
    # CSV 読み込み
    # -------------------------------------------------------
    df = pd.read_csv(input_csv)

    parcel_cols = [c for c in df.columns if c.startswith("parcel_")]
    valid_subs = df["subject"].unique()
    subjects = len(valid_subs)

    n_TR = df["TR"].max() + 1
    n_parcels = len(parcel_cols)

    # -------------------------------------------------------
    #  (subjects, TR, parcels) に再構成
    # -------------------------------------------------------
    all_pred = np.zeros((subjects, n_TR, n_parcels))

    for i, subname in enumerate(valid_subs):
        sub_df = df[df["subject"] == subname].sort_values("TR")
        all_pred[i] = sub_df[parcel_cols].values

    # -------------------------------------------------------
    # あなたが以前使った正規化：reshape → MinMaxScaler → reshape back
    # subjectsごとにまとめてスケール
    # -------------------------------------------------------
    all_pred_2d = all_pred.reshape(subjects, n_TR * n_parcels)

    scaler = MinMaxScaler()
    all_pred_scaled_2d = scaler.fit_transform(all_pred_2d)

    all_pred_scaled = all_pred_scaled_2d.reshape(subjects, n_TR, n_parcels)

    # -------------------------------------------------------
    # DataFrameに戻す
    # -------------------------------------------------------
    rows = []
    scaled_values = []

    for i, sub in enumerate(valid_subs):
        for tr in range(n_TR):
            rows.append([sub, tr])
            scaled_values.append(all_pred_scaled[i, tr])

    df_out = (
        pd.DataFrame(rows, columns=["subject", "TR"])
        .join(pd.DataFrame(scaled_values, columns=parcel_cols))
    )

    # -------------------------------------------------------
    # 保存
    # -------------------------------------------------------
    df_out.to_csv(output_csv, index=False, encoding="utf-8-sig")
    print(f"   ✔ 保存しました: {output_csv}")

print("\n🎉 すべてのモデルのスケーリングが完了しました！")
```
