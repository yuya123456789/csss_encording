```python
from os.path import exists
import pandas as pd
import numpy as np
def construct_predictors(transcript_df, n_features, stim_dur, tr=1.5):

    # Find total number of TRs
    stim_trs = np.ceil(stim_dur / tr)

    # Add column to transcript with TR indices
    transcript_df['TR'] = transcript_df.onset.divide(tr).apply(np.floor).apply(int)
    
    # Compile the words within each TR
    words_per_tr = transcript_df.groupby('TR')['word'].apply(list)
    
    # Average the embeddings within each TR
    embeddings_per_tr = transcript_df.groupby('TR')['embedding'].mean()
    
    # Loop through TRs
    words_trs = []
    embeddings_trs = []
    for t in np.arange(stim_trs):
        if t in words_per_tr:
            words_trs.append(words_per_tr[t])
    
            # Fill in empty TRs with zero vectors
            if embeddings_per_tr[t] is not np.nan:
                embeddings_trs.append(embeddings_per_tr[t])
            else:
                embeddings_trs.append(np.zeros(n_features))
        else:
            words_trs.append([])
            embeddings_trs.append(np.zeros(n_features))
    
    embeddings = np.vstack(embeddings_trs)
    return embeddings

import pickle
transcript_f = '../emb/piemanpni_gpt2.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gpt2 = pickle.load(f)

transcript_f = '../emb/piemanpni_gte.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gte = pickle.load(f)

transcript_f = '../emb/piemanpni_llama.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_llama = pickle.load(f)
        
transcript_f = '../emb/piemanpni_w2v.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_w2v = pickle.load(f)
        
transcript_f = '../emb/piemanpni_llama3.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_llama = pickle.load(f)

transcript_f = '../emb/piemanpni_gpt_oss.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gpt_oss = pickle.load(f)

from transformers import AutoTokenizer, AutoModel
import pickle
import os
OUT_DIR = "../emb/"
transcript_gpt2 = pd.read_pickle(os.path.join(OUT_DIR, "../emb/piemanpni_gpt2.pkl"))

# まず Word2Vec の“古い embedding 列”を削除
if "embedding" in transcript_gpt2.columns:
    transcript_gpt2 = transcript_gpt2.drop(columns=["embedding"])

# GPT-2 の列を "embedding" に統一
transcript_gpt2 = transcript_gpt2.rename(columns={"embedding_gpt2": "embedding"})

print("GPT2 embedding shape (first row):", transcript_gpt2["embedding"].iloc[0].shape)

transcript_gte = pd.read_pickle(os.path.join(OUT_DIR, "../emb/piemanpni_gte.pkl"))
if "embedding" in transcript_gte.columns:
    transcript_gte = transcript_gte.drop(columns=["embedding"])
transcript_gte = transcript_gte.rename(columns={"embedding_gte": "embedding"})

transcript_llama = pd.read_pickle(os.path.join(OUT_DIR, "../emb/piemanpni_llama.pkl"))
if "embedding" in transcript_llama.columns:
    transcript_llama = transcript_llama.drop(columns=["embedding"])
transcript_llama = transcript_llama.rename(columns={"embedding_llama": "embedding"})

transcript_llama3 = pd.read_pickle(os.path.join(OUT_DIR, "../emb/piemanpni_llama3.pkl"))
if "embedding" in transcript_llama3.columns:
    transcript_llama3 = transcript_llama3.drop(columns=["embedding"])
transcript_llama3 = transcript_llama3.rename(columns={"embedding_llama3": "embedding"})
print("llama3 embedding shape (first row):", transcript_llama3["embedding"].iloc[0].shape)

transcript_gpt_oss = pd.read_pickle(os.path.join(OUT_DIR, "../emb/piemanpni_gpt_oss.pkl"))
if "embedding" in transcript_gpt_oss.columns:
    transcript_gpt_oss = transcript_gpt_oss.drop(columns=["embedding"])
transcript_gpt_oss = transcript_gpt_oss.rename(columns={"embedding_gpt_oss": "embedding"})

X_w2v = construct_predictors(transcript_w2v, 300, 400, tr=1.5)

# GPT-2 embeddings are 768-dimensional
X_gpt2 = construct_predictors(transcript_gpt2, 768, 400, tr=1.5)

X_gpt_oss = construct_predictors(transcript_gpt_oss, 2880, 400, tr=1.5)

X_gte = construct_predictors(transcript_gte, 1024, 400, tr=1.5)

X_llama = construct_predictors(transcript_llama, 4096, 400, tr=1.5)

X_llama3 = construct_predictors(transcript_llama3, 4096, 400, tr=1.5)
```

```python
import numpy as np
import pandas as pd
from brainiak.isc import isc
from voxelwise_tutorials.delayer import Delayer
from himalaya.scoring import correlation_score
from nilearn.input_data import NiftiLabelsMasker
from nilearn.datasets import fetch_atlas_schaefer_2018
from himalaya.kernel_ridge import KernelRidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
import os

# ============================================================
# 設定
# ============================================================
task_train = "piemanpni"     # 学習に使う物語
task_test  =  "bronx" # 評価に使う別物語
base_dir = os.getcwd()
label_path = os.path.join(os.path.dirname(base_dir),
                          'atlas_label/Schaefer2018_400Parcels_7Networks_order_info.txt')

model_list = ["w2v", "gpt2", "gpt_oss", "llama", "llama3", "gte"]

# ============================================================
# 7ネットワークラベル読み込み
# ============================================================
def load_network_labels_7networks(label_fn):
    networks = []
    with open(label_fn, 'r') as f:
        for line in f:
            if '7Networks_' in line:
                network_name = line.split('_')[2]
                networks.append(network_name)
    unique_network_labels = sorted(set(networks), key=networks.index)
    network_indices = [
        int(np.median([i for i, n in enumerate(networks) if n == network]))
        for network in unique_network_labels
    ]
    return networks, unique_network_labels, network_indices

networks, unique_networks, indices = load_network_labels_7networks(label_path)

# ============================================================
# Atlas
# ============================================================
n_parcels = 400
atlas = fetch_atlas_schaefer_2018(n_rois=n_parcels, yeo_networks=17, resolution_mm=2)
masker = NiftiLabelsMasker(atlas.maps, labels=atlas.labels, standardize=True)

# ============================================================
# 脳データ読み込み（共通）
# ============================================================
df_all = pd.read_csv(f"../parcels_csv/{task_train}_clean-parcels_all_subjects.csv")
valid_subs = sorted(df_all["sub"].unique())
n_TR = df_all["TR"].max() + 1
parcel_cols = [c for c in df_all.columns if c.startswith("parcel_")]
n_parcels = len(parcel_cols)

all_ts = np.zeros((len(valid_subs), n_TR, n_parcels))
for i, subname in enumerate(valid_subs):
    sub_df = df_all[df_all["sub"] == subname].sort_values("TR")
    all_ts[i] = sub_df[parcel_cols].values

df_test = pd.read_csv(f"../parcels_csv/{task_test}_clean-parcels_all_subjects.csv")
valid_subs_test = sorted(df_test["sub"].unique())
n_TR_test = df_test["TR"].max() + 1

all_ts_test = np.zeros((len(valid_subs_test), n_TR_test, n_parcels))
for i, subname in enumerate(valid_subs_test):
    sub_df = df_test[df_test["sub"] == subname].sort_values("TR")
    all_ts_test[i] = sub_df[parcel_cols].values

print("✅ all_ts shape =", all_ts.shape)

# ============================================================
# モデル設定（共通）
# ============================================================
outer_cv = KFold(n_splits=2)
inner_cv = KFold(n_splits=5)
scaler = StandardScaler(with_mean=True, with_std=True)
delayer = Delayer(delays=[2, 3, 4, 5])
# alphas = np.logspace(1, 10, 10)
alphas = np.logspace(-3,4,20)
ridge = KernelRidgeCV(alphas=alphas, cv=inner_cv)
pipeline = make_pipeline(scaler, delayer, ridge)

pred_dir = "../predicted_all_csv/"
os.makedirs(pred_dir, exist_ok=True)

# ============================================================
# すべてのモデルでループ
# ============================================================
for m_name in model_list:
    print("\n===================================================")
    print(f"  モデル処理中: {m_name}")
    print("===================================================\n")

    # ====== 埋め込み読み込み ======
    try:
        X = globals()[f"X_{m_name}"]
        print(f"✅ X_{m_name} を使用")
    except KeyError:
        print(f"❌ X_{m_name} が未定義、スキップします")
        continue

    try:
        X_test = globals()[f"X_bronx_{m_name}"]
        print(f"✅ X_bronx_{m_name} を使用")
    except KeyError:
        print(f"❌ X_bronx_{m_name} が未定義、スキップします")
        continue
    n_test = len(X_test)
    mid = n_test // 2

    X_test_first  = X_test[:mid]
    X_test_second = X_test[mid:]
    # ============================================================
    # 各モデル用の保存リスト
    # ============================================================
    all_Y_train = []     # piemanpni → CV
    all_Y_test  = []     # piemanpni → bronx

    # ============================================================
    # 被験者ループ
    # ============================================================
    for idx_self, subname in enumerate(valid_subs):
        print(f"--- Subject: {subname} ({idx_self+1}/{len(valid_subs)}) ---")

        # ====== CV prediction (piemanpni) ======
        Y_parcels = all_ts[idx_self]
        preds = []
        preds_test = []
        for i, (train, test) in enumerate(outer_cv.split(Y_parcels)):
            pipeline.fit(X[train], Y_parcels[train])
            Y_pred = pipeline.predict(X[test])
            preds.append(Y_pred)
            if i % 2 == 0:
                preds_test.append(pipeline.predict(X_test_first))
            else:
                preds_test.append(pipeline.predict(X_test_second))

        preds_test=np.vstack(preds_test)
        preds = np.vstack(preds)
        all_Y_train.append(preds)
        all_Y_test.append(preds_test)
            
        # preds = np.vstack(preds)
        # all_Y_train.append(preds)
        
        
        # ====== pieman prediction ======
        idx_test = valid_subs_test.index(subname)
        
        Y_parcels_test = all_ts_test[idx_test]
        # pred_neo = pipeline.predict(X_test)
        # all_Y_test.append(pred_neo)

    # ============================================================
    # CSV 保存
    # ============================================================
    # ---- piemanpni CV ----
    Y_piemanpni = np.stack(all_Y_train, axis=0)
    n_subs, n_tr, n_parc = Y_piemanpni.shape
    Y_piemanpni_2d = Y_piemanpni.reshape(n_subs * n_tr, n_parc)

    rows = [(sub, tr) for sub in valid_subs for tr in range(n_tr)]
    df_index = pd.DataFrame(rows, columns=["subject", "TR"])
    df_pred = pd.concat([df_index,
                         pd.DataFrame(Y_piemanpni_2d, columns=[f"parcel_{i}" for i in range(n_parc)])],
                        axis=1)

    out1 = os.path.join(pred_dir, f"{m_name}_piemanpni_clean_predicted.csv")
    df_pred.to_csv(out1, index=False, encoding="utf-8-sig")

    # ---- pieman prediction ----
    Y_pieman = np.stack(all_Y_test, axis=0)
    n_subs2, n_tr2, n_parc2 = Y_pieman.shape
    Y_pieman_2d = Y_pieman.reshape(n_subs2 * n_tr2, n_parc2)

    rows2 = [(sub, tr) for sub in valid_subs for tr in range(n_tr2)]
    df_index2 = pd.DataFrame(rows2, columns=["subject", "TR"])
    df_pred2 = pd.concat([df_index2,
                          pd.DataFrame(Y_pieman_2d, columns=[f"parcel_{i}" for i in range(n_parc2)])],
                         axis=1)

    out2 = os.path.join(pred_dir, f"{m_name}_piemanpni_to_bronx_clean_predicted.csv")
    df_pred2.to_csv(out2, index=False, encoding="utf-8-sig")

    print(f"✔ 保存完了: {out1}")
    print(f"✔ 保存完了: {out2}")

print("\n🎉 全モデルの処理が完了しました！")
```
