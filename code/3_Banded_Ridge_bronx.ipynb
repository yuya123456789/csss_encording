```python
from os.path import exists
import numpy as np
import pandas as pd
def construct_predictors(transcript_df, n_features, stim_dur, tr=1.5):

    # Find total number of TRs
    stim_trs = np.ceil(stim_dur / tr)

    # Add column to transcript with TR indices
    transcript_df['TR'] = transcript_df.onset.divide(tr).apply(np.floor).apply(int)
    
    # Compile the words within each TR
    words_per_tr = transcript_df.groupby('TR')['word'].apply(list)
    
    # Average the embeddings within each TR
    embeddings_per_tr = transcript_df.groupby('TR')['embedding'].mean()
    
    # Loop through TRs
    words_trs = []
    embeddings_trs = []
    for t in np.arange(stim_trs):
        if t in words_per_tr:
            words_trs.append(words_per_tr[t])
    
            # Fill in empty TRs with zero vectors
            if embeddings_per_tr[t] is not np.nan:
                embeddings_trs.append(embeddings_per_tr[t])
            else:
                embeddings_trs.append(np.zeros(n_features))
        else:
            words_trs.append([])
            embeddings_trs.append(np.zeros(n_features))
    
    embeddings = np.vstack(embeddings_trs)
    return embeddings

import pickle
transcript_f = '../emb/bronx_gpt2.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gpt2 = pickle.load(f)

transcript_f = '../emb/bronx_gte.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gte = pickle.load(f)

transcript_f = '../emb/bronx_llama.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_llama = pickle.load(f)
        
transcript_f = '../emb/bronx_w2v.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_w2v = pickle.load(f)
        
transcript_f = '../emb/bronx_llama3.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_llama = pickle.load(f)

transcript_f = '../emb/bronx_gpt_oss.pkl'
if exists(transcript_f):
    with open(transcript_f, 'rb') as f:
        transcript_gpt_oss = pickle.load(f)

from transformers import AutoTokenizer, AutoModel
import pickle
import os
OUT_DIR = "../emb/"
transcript_gpt2 = pd.read_pickle(os.path.join(OUT_DIR, "../emb/bronx_gpt2.pkl"))

# まず Word2Vec の“古い embedding 列”を削除
if "embedding" in transcript_gpt2.columns:
    transcript_gpt2 = transcript_gpt2.drop(columns=["embedding"])

# GPT-2 の列を "embedding" に統一
transcript_gpt2 = transcript_gpt2.rename(columns={"embedding_gpt2": "embedding"})

print("GPT2 embedding shape (first row):", transcript_gpt2["embedding"].iloc[0].shape)

transcript_gte = pd.read_pickle(os.path.join(OUT_DIR, "../emb/bronx_gte.pkl"))
if "embedding" in transcript_gte.columns:
    transcript_gte = transcript_gte.drop(columns=["embedding"])
transcript_gte = transcript_gte.rename(columns={"embedding_gte": "embedding"})

transcript_llama = pd.read_pickle(os.path.join(OUT_DIR, "../emb/bronx_llama.pkl"))
if "embedding" in transcript_llama.columns:
    transcript_llama = transcript_llama.drop(columns=["embedding"])
transcript_llama = transcript_llama.rename(columns={"embedding_llama": "embedding"})

transcript_llama3 = pd.read_pickle(os.path.join(OUT_DIR, "../emb/bronx_llama3.pkl"))
if "embedding" in transcript_llama3.columns:
    transcript_llama3 = transcript_llama3.drop(columns=["embedding"])
transcript_llama3 = transcript_llama3.rename(columns={"embedding_llama3": "embedding"})
print("llama3 embedding shape (first row):", transcript_llama3["embedding"].iloc[0].shape)

transcript_gpt_oss = pd.read_pickle(os.path.join(OUT_DIR, "../emb/bronx_gpt_oss.pkl"))
if "embedding" in transcript_gpt_oss.columns:
    transcript_gpt_oss = transcript_gpt_oss.drop(columns=["embedding"])
transcript_gpt_oss = transcript_gpt_oss.rename(columns={"embedding_gpt_oss": "embedding"})

X_w2v = construct_predictors(transcript_w2v, 300, 535.5, tr=1.5)

# GPT-2 embeddings are 768-dimensional
X_gpt2 = construct_predictors(transcript_gpt2, 768, 535.5, tr=1.5)

X_gpt_oss = construct_predictors(transcript_gpt_oss, 2880, 535.5, tr=1.5)

X_gte = construct_predictors(transcript_gte, 1024, 535.5, tr=1.5)

X_llama = construct_predictors(transcript_llama, 4096, 535.5, tr=1.5)

X_llama3 = construct_predictors(transcript_llama3, 4096, 535.5, tr=1.5)
```
